```python
# path = '/Users/yiqinhuang/NLP/news-deep-researcher/notebooks/retriever_validation_k_100.json'
# path = '/Users/yiqinhuang/NLP/news-deep-researcher/notebooks/retriever_validation.json'
# path = '/Users/yiqinhuang/NLP/news-deep-researcher/notebooks/retriever_validation_4.json'
# path = '/Users/yiqinhuang/NLP/news-deep-researcher/notebooks/retriever_validation_6.json'
#path = '/Users/yiqinhuang/NLP/news-deep-researcher/notebooks/retriever_output_query_improved.json'
path = '/Users/yiqinhuang/NLP/news-deep-researcher/notebooks/retriever_validation_claim.json'
#path = '/Users/yiqinhuang/NLP/news-deep-researcher/notebooks/retriever_validation_improved_1.json'

with open(path, 'r', encoding='utf-8') as f:
    datapoints = json.load(f)

for point in datapoints:
    print("TARGET : ")
    pprint(point['input_data']['target'])
    print("\n")
    for query_dict in point['llm_response']['queries_only']:
        print("QUERY: ")
        pprint(query_dict['query'])
        print("\n")
        print("RETRIEVAL: ")
        for doc in query_dict['retrieval']:
            print(doc['page_content'])
            print("\n")
        
        print("\n")
```

```python
total_queries = 0
successful_queries = 0
total_datapoints = 0
successful_datapoints = 0
for point in datapoints:
    total_datapoints += 1
    success = False
    target = point['input_data']['target']
    for query_dict in point['llm_response']['queries_only']:
        total_queries += 1
        for source in query_dict['retrieval']:
            source_dict = source['metadata']['source']
            if target == source_dict:
                successful_queries += 1
                success = True
                query_dict['success'] = True
    
    if success:
        successful_datapoints += 1
    # else:
    #     pprint(point['input_data']['target'])
    #     print("Failed Queries: ")
    #     pprint([p['query'] for p in point['llm_response']['queries_only']])
    #     print('\n')

print("total_queries: ", total_queries)
print("successful_queries: ", successful_queries)
print("total_datapoints: ", total_datapoints)
print("successful_datapoints: ", successful_datapoints)
```
